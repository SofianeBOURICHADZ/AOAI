```python
import pandas as pd
import numpy as np
df = pd.read_csv("USA_Housing.csv")
df = df.drop(labels="Address", axis = 1)
df
```

```python
trainX = np.c_[np.ones(3000), df[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Avg. Area Number of Bedrooms", "Area Population"]].iloc[:3000,:].to_numpy()]
trainY = df["Price"].iloc[:3000].to_numpy()

theta = np.linalg.inv(trainX.T @ trainX) @ trainX.T @ trainY
```

```python
theta
```

```python
testX = np.c_[np.ones(2000), df[["Avg. Area Income", "Avg. Area House Age", "Avg. Area Number of Rooms", "Avg. Area Number of Bedrooms", "Area Population"]].iloc[3000:,:].to_numpy()]
testY = df["Price"].iloc[3000:].to_numpy()

predictY = testX @ theta

cost = np.sum((predictY - testY) ** 2) / (2 * len(testY))
rmse = np.sqrt(np.mean((predictY - testY) ** 2))

print(f"{rmse}         {cost}")
```
