```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv("Salary_Data.csv").iloc[:20,:]
x = data["YearsExperience"].to_numpy()
y = data["Salary"].to_numpy()

X = np.c_[np.ones(x.shape[0], dtype = int), x]

theta = np.linalg.inv(X.T @ X) @ X.T @ y

test = pd.read_csv("Salary_Data.csv").iloc[20:,:]
testx = test["YearsExperience"].to_numpy()
testX = np.c_[np.ones(testx.shape[0], dtype = int), testx]
testy = test["Salary"].to_numpy()

predictY = testX @ theta

cost = np.sum((predictY - testy)**2) / (2*testX.shape[0])
predictY
plt.plot(predictY,testy)
plt.show()



```
